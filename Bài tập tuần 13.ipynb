```python
#W11A1
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        n = f.read().strip()   
    reversed_n = n[::-1]      
    return int(reversed_n)    
```

```python
#W11A4
def areAnagrams(path1: str, path2: str) -> bool:
    def normalize(path):
        with open(path, 'r', encoding='utf-8') as f:
            s = f.read().lower()
        count = {}
        for ch in s:
            if ch.isalpha():
                count[ch] = count.get(ch, 0) + 1
        return count

    return normalize(path1) == normalize(path2)
```

```python
#W11A5
def count(path: str, character: str):
    with open(path, 'r', encoding='utf-8') as f:
        sentence = f.read()
    cnt = 0
    for ch in sentence:
        if ch == character:
            cnt += 1
    return cnt
```

```python
#W11A6
def countCharacter(path: str, letter: str) -> int:
    with open(path, 'r', encoding='utf-8') as f:
        text = f.read().lower()
    return text.count(letter.lower())
```

```python
#W11A12
def maximumProduct(path: str) -> int:
    with open(path, 'r', encoding='utf-8') as f:
        n = int(f.readline().strip())
        nums = list(map(int, f.readline().split()))
    nums.sort()
    product1 = nums[-1] * nums[-2] * nums[-3]   
    product2 = nums[0] * nums[1] * nums[-1]      
    return max(product1, product2)
```

```python
#W11A13
def moveZeroes(fileName: str) -> list:
    with open(fileName, 'r', encoding='utf-8') as f:
        n = int(f.readline().strip())
        nums = list(map(int, f.readline().split()))
    result = []
    zero_count = 0
    for x in nums:
        if x == 0:
            zero_count += 1
        else:
            result.append(x)

    result.extend([0] * zero_count)
    return result
```

```python
#W11A14
def averageTime(path: str) -> float:
    total = 0.0
    count = 0
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    for line in lines[1:]:
        parts = line.strip().split(',')
        time = float(parts[1])
        total += time
        count += 1
    if count == 0:
        return 0.00

    return round(total / count, 2)
```

```python
#W11A15
def findMovies(path: str) -> list[str] | str:
    result = []
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    for line in lines[1:]:
        parts = [p.strip() for p in line.split(',')]
        movie = parts[1]
        description = parts[2].lower()
        rating = float(parts[3])
        if rating > 8.0 and description != "boring":
            result.append(movie)
    if not result:
        return "All are boring!"

    return result
```

```python
#W11A16
import math
def printPerimeters(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip() != ""]
    i = 0
    while i < len(lines):
        shape_type = lines[i].split(":")[1].strip()
        params = lines[i + 1]
        i += 2
        if shape_type == "SQUARE":
            side = float(params.split(":")[1])
            perimeter = 4 * side
        elif shape_type == "RECTANGLE":
            parts = params.split()
            width = float(parts[1])
            height = float(parts[3])
            perimeter = 2 * (width + height)
        elif shape_type == "CIRCLE":
            radius = float(params.split(":")[1])
            perimeter = 2 * math.pi * radius

        print(f"{perimeter:.2f}")
```

```python
#W11A17
def productExceptSelf(path: str) -> list[int]:
    with open(path, 'r', encoding='utf-8') as f:
        nums = list(map(int, f.read().split()))
    n = len(nums)
    left = [1] * n
    right = [1] * n
    result = [0] * n
    for i in range(1, n):
        left[i] = left[i - 1] * nums[i - 1]
    for i in range(n - 2, -1, -1):
        right[i] = right[i + 1] * nums[i + 1]
    for i in range(n):
        result[i] = left[i] * right[i]

    print(result)
    return result
```

```python
#W11A18
import math
def estimatedTime(path: str) -> float:
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    src_parts = lines[0].split()
    x1 = float(src_parts[2])
    y1 = float(src_parts[4])
    des_parts = lines[1].split()
    x2 = float(des_parts[2])
    y2 = float(des_parts[4])
    vel_parts = lines[2].split()
    velocity = float(vel_parts[1])
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    time = distance / velocity

    return round(time, 2)
```

```python
#W11A20
def imageSmoother(path: str):
    with open(path, "r") as f:
        lines = f.read().strip().splitlines()
    M = int(lines[0])
    img = []
    for i in range(1, M + 1):
        img.append(list(map(int, lines[i].split())))
    result = [[0] * M for _ in range(M)]
    for i in range(M):
        for j in range(M):
            total = 0
            count = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    ni = i + dx
                    nj = j + dy
                    if 0 <= ni < M and 0 <= nj < M:
                        total += img[ni][nj]
                        count += 1
            result[i][j] = total // count

    for row in result:
        print(*row)
```
