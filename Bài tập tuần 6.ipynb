```python
#W6A9
n1 = int(input())
dict1 = {}
for _ in range(n1):
    k, v = input().split()
    dict1[k] = int(v)

n2 = int(input())
dict2 = {}
for _ in range(n2):
    k, v = input().split()
    dict2[k] = int(v)


result = {}

for k in dict1:
    result[k] = dict1[k]

for k in dict2:
    if k in result:
        result[k] += dict2[k]
    else:
        result[k] = dict2[k]

for k in sorted(result):
    print(k, result[k])
```

```python
#W6A10
a = list(map(int, input().split()))
k = int(input())
result = []
n = len(a)
for i in range(n - 1):
    for j in range(i + 1, n):
        if a[i] + a[j] == k:
            pair = (a[i], a[j])
            if pair not in result:
                result.append(pair)

result.sort(key=lambda x: x[0])

print(result)
```

```python
#W6A12
a = list(map(int, input().split()))

freq = {}

for x in a:
    if x in freq:
        freq[x] += 1
    else:
        freq[x] = 1

max_count = 0
result = None

for x in freq:
    if freq[x] > max_count:
        max_count = freq[x]
        result = x
    elif freq[x] == max_count and x < result:
        result = x

print(result)
```
