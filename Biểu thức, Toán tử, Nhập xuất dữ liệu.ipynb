```python
a=input('nhập kí tự: ')
if len(a)!=1:
    print('vui lòng nhập đúng 1 kí tự')
elif a.islower():
    print('mã unicode là: ',ord(a))
    print('chữ hoa tương ứng là:',a.upper())
else:
    print("vui lòng nhập kí tự thường")
```
