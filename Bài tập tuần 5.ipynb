```python
#W5A1
def Max_Of_Two(a: int, b: int) -> int:
    if a >= b:
        return a
    else:
        return b
```

```python
#W5A2
from typing import Tuple
def swap(a: int, b: int) -> Tuple[int, int]:
    a, b = b, a
    return a, b
```

```python
#W5A3
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
#W5A4
def is_perfect(n :int) -> bool:
    if n <= 1:
        return False

    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i

    return tong == n
```

```python
#W5A5
def first_pos(a, k):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1
```

```python
#W5A6
def giai_thua(n ):
    if n < 0:
        return -1
    else:
        gt=1
        for i in range(1,n+1):
            gt *= i
        return gt
    
```

```python
#W5A7
def calculator(num1, num2, operat):
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return None  
        result = num1 / num2
    else:
        return None  

    return round(result, 2)
```

```python
#W5A8
def hamming_distance(x, y):
    xor = x ^ y
    count = 0
    while xor > 0:
        count += xor & 1
        xor >>= 1
    return count
```

```python
#W5A9
def sum_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
```

```python
#W5A10
def is_isomorphic(a, b):
    map_ab = {}
    map_ba = {}

    for i in range(len(a)):
        c1 = a[i]
        c2 = b[i]

        if c1 in map_ab:
            if map_ab[c1] != c2:
                return False
        else:
            map_ab[c1] = c2

        if c2 in map_ba:
            if map_ba[c2] != c1:
                return False
        else:
            map_ba[c2] = c1

    return True
```
