```python
#W7A16
def make_beautiful_array(arr):
    freq = {}
    for x in arr:
        if x in freq:
            freq[x] += 1
        else:
            freq[x] = 1

    deletions = 0  
    result = []    
    for x in freq:
        count = freq[x]
        if count == x:
            result += [x] * count
        elif count > x:
            result += [x] * x
            deletions += count - x
        else:  
            deletions += count

    return result, deletions
arr1 = [2, 2, 3, 3, 3, 1]
arr2 = [2, 2, 2, 1]

print(make_beautiful_array(arr1))  
print(make_beautiful_array(arr2))  
```

```python
#W7A17
def majority_element(arr):
    n = len(arr)
    if n == 0:
        return None
    candidate = None
    count = 0
    for x in arr:
        if count == 0:
            candidate = x
            count = 1
        elif x == candidate:
            count += 1
        else:
            count -= 1
    cnt = 0
    for x in arr:
        if x == candidate:
            cnt += 1

    if cnt > n // 2:
        return candidate
    else:
        return None
```

```python
#W7A18
def conditional_sort(arr):
    even = []
    odd = []
    for x in arr:
        if x % 2 == 0:
            even.append(x)
        else:
            odd.append(x)
    even.sort()              
    odd.sort(reverse=True)  
    return even + odd
```

```python
#W7A19
def merge_sorted_arrays(a, b):
    n, m = len(a), len(b)
    i = j = 0
    result = []
    if n == 0:
        return b[:]
    if m == 0:
        return a[:]
    ascending = a[0] <= a[-1]
    while i < n and j < m:
        if ascending:
            if a[i] <= b[j]:
                result.append(a[i])
                i += 1
            else:
                result.append(b[j])
                j += 1
        else:  
            if a[i] >= b[j]:
                result.append(a[i])
                i += 1
            else:
                result.append(b[j])
                j += 1
    while i < n:
        result.append(a[i])
        i += 1
    while j < m:
        result.append(b[j])
        j += 1
    return result
    
```

```python
#W7A25
s = input().strip()  
k = int(input())
stack = []
n = len(s)
for i in range(n):
    d = s[i]
    while (stack and 
           stack[-1] < d and 
           len(stack) + (n - i) > k):
        stack.pop()
    if len(stack) < k:
        stack.append(d)

print("".join(stack))
```

```python
#W7A28
def is_composite(x):
    if x <= 3:
        return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return True
    return False
N = int(input())
best_start = 0
best_len = 0
current_start = 0
current_len = 0
for i in range(1, N + 1):
    if is_composite(i):
        if current_len == 0:
            current_start = i
        current_len += 1
    else:
        if current_len > best_len:
            best_len = current_len
            best_start = current_start
        current_len = 0
if current_len > best_len:
    best_len = current_len
    best_start = current_start
if best_len == 0:
    print("Không có dãy hợp số")
else:
    for i in range(best_start, best_start + best_len):
        print(i, end=" ")
```

```python
#W7A30
a = list(map(int, input().split()))
n = len(a)
dp = [1] * n
prev = [-1] * n
for i in range(n):
    for j in range(i):
        if a[j] < a[i] and dp[j] + 1 > dp[i]:
            dp[i] = dp[j] + 1
            prev[i] = j
max_len = max(dp)
end = dp.index(max_len)  
lis = []
while end != -1:
    lis.append(a[end])
    end = prev[end]
lis.reverse()
for x in lis:
    print(x, end=" ")
```
