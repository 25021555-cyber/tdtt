```python
#bài 8
a = int(input("Nhập số nguyên dương a: "))
b = int(input("Nhập số nguyên dương b: "))
if a <= 0 or b <= 0:
    print("a và b phải là số nguyên dương.")
else:
    print("Các ước dương chung của", a, "và", b, "là:")
    limit = min(a, b)
    for i in range(1, limit + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=" ")
```

```python
#bài 12
n = int(input("Nhập số nguyên dương n: "))
S = 0
k = 0
while True:
    k += 1
    S += k
    if S >= n:
        k -= 1  
        break

print("k lớn nhất sao cho S(k) < n là:", k)
```

```python
#bài 13
A = float(input("Nhập A: "))
if A < 0:
    print("n nhỏ nhất là 1")
else:
    s = 0.0
    n = 0
    while s <= A:
        n += 1
        s += 1.0 / n
    print("n nhỏ nhất sao cho 1 + 1/2 + ... + 1/n >", A, "là:", n)
```

```python
#bài 14
print("Nhập các số nguyên, nhập -1 để dừng:")
x = int(input())
if x == -1:
    print("Không có số nào được nhập!")
else:
    min_val = x
    max_val = x

    while True:
        x = int(input())
        if x == -1:
            break

        if x < min_val:
            min_val = x
        if x > max_val:
            max_val = x

    print("Số nhỏ nhất:", min_val)
    print("Số lớn nhất:", max_val)
```

```python
#bài 16
n = input("Nhập số nguyên dương n: ")

chan = 0
le = 0

for ch in n:
    digit = int(ch)
    if digit % 2 == 0:
        chan += 1
    else:
        le += 1

print("Số chữ số chẵn:", chan)
print("Số chữ số lẻ:", le)
```

```python
#bài 18
n = int(input("Nhập số nguyên dương n: "))

if n <= 0:
    print("n phải là số nguyên dương")
else:
    temp = n
    while temp % 3 == 0:
        temp //= 3

    if temp == 1:
        print(n, "là số dạng 3^k")
    else:
        print(n, "không phải số dạng 3^k")
```

```python
#bài 19
A = int(input("Nhập số nguyên dương A: "))
f1 = 1
f2 = 1
if A < 1:
    print("Không có số Fibonacci nào ≤ A")
else:
    while f2 <= A:
        f1, f2 = f2, f1 + f2

    print("Số Fibonacci lớn nhất không vượt quá", A, "là:", f1)
```
