```python
#W4A6
def is_prime(n):
    if n < 2:
        return False
    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return True

print("Nhập các bộ test (a b). Nhập dòng trống để dừng:")

while True:
    try:
        line = input().strip()
        if line == "":
            break      

        a, b = map(int, line.split())
        tong = 0
        
        for x in range(a, b + 1):
            if is_prime(x):
                tong += x
        
        print("Tổng các số nguyên tố trong đoạn [{}, {}] là: {}".format(a, b, tong))

    except EOFError:
        break
```

```python
#W4A7
n = int(input("Nhập n (n ≥ 2): "))

temp = n
largest_prime = 0

d = 2
while d * d <= temp:
    while temp % d == 0:
        largest_prime = d    
        temp //= d
    d += 1

if temp > 1:
    largest_prime = temp

print("Ước số nguyên tố lớn nhất của", n, "là:", largest_prime)
```

```python
#W4A8
n = int(input("Nhập số nguyên dương n: "))
steps = 0
while True:
    temp = n
    rev = 0
    while temp > 0:
        rev = rev * 10 + temp % 10
        temp //= 10
    if n == rev:
        break

    n = n + rev
    steps += 1

print("Số bước thực hiện:", steps)
print("Giá trị palindrome thu được:", n)
```

```python
#W4A9
n = int(input("Nhập n: "))
found = False
for i in range(int(n**0.5) + 1):       
    sq = i * i
    temp = sq
    used = [0] * 10       
    ok = True
    if sq == 0:
        used[0] = 1
    else:
        while temp > 0:
            d = temp % 10
            if used[d] == 1:   
                ok = False
                break
            used[d] = 1
            temp //= 10

    if ok:
        print(sq, end=" ")
        found = True

if not found:
    print("(no number)")
```

```python
#W4A10
n = int(input("Nhập n: "))
max_len = 0
best_x = 1
for x in range(1, n + 1):
    t = x
    length = 1

    while t != 1:
        if t % 2 == 0:
            t //= 2
        else:
            t = 3 * t + 1
        length += 1

    if length > max_len or (length == max_len and x < best_x):
        max_len = length
        best_x = x

print(best_x, max_len)
```

```python
#W4A11
def dem_uoc_chan_toi_uu(n):
    if n <= 0 or n % 2 != 0:
        return 0 
    n = n // 2
    dem = 0
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            dem += 1
            if i != n // i:
                dem += 1
    return dem

n = 24
print(f"Số ước chẵn của {n}: {dem_uoc_chan_toi_uu(n)}")
```

```python
#W4A14
m = int(input("Nhập m: "))
n = int(input("Nhập n: "))

gcd = 1
for i in range(1, min(m, n) + 1):
    if m % i == 0 and n % i == 0:
        gcd = i

print("Ước chung lớn nhất là:", gcd)
```
