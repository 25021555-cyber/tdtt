```python
#bài 5
import math
a,b,c=map(float,input("Nhập 3 số a,b,c: ").split())
if a==0:
    if b==0:
        print("không phải phương trình")
    else:
        print("phương trình có nghiệm x: ", -c/b)
else:
    delta=b**2-4*a*c
    if delta<0:
        print("phương trình vô nghiệm")
    elif delta==0:
        x=-b/(2*a)
        print("phương trình có nghiệm kép x1=x2=",x)
    else:
        x1=(-b+math.sqrt(delta))/(2*a)
        x2=(-b-math.sqrt(delta))/(2*a)
        print("phương trình có 2 nghiệm phân biệt:")
        print("x1=",x1)
        print("x2=",x2)
```

```python
#bài 8
a,b,m=map(float, input("Nhập a,b,m: ").split())
c,d,n=map(float, input("Nhập c,d,n: ").split())
if a*d==b*c:
    if m*d==b*n and a*n==c*m:
        print("hệ phương trình có vô số nghiệm")
    else:
        print("hệ phương trình vô nghiệm")
else:
    x=(m*d - b*n)/(a*d - b*c)
    y=(a*n - c*m)/(a*d - b*c)
    print("hệ phương trình có nghiệm duy nhất:")
    print("x=",x)
    print("y=",y)
```
