```python
#W9A1
def merge_sorted_lists(a, b):
    i = j = 0
    result = []
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            result.append(a[i])
            i += 1
        else:
            result.append(b[j])
            j += 1
    while i < len(a):
        result.append(a[i])
        i += 1
    while j < len(b):
        result.append(b[j])
        j += 1
    return result
```

```python
#W9A2
def count_ge(matrix, mid):
    n = len(matrix)
    m = len(matrix[0])
    i = n - 1
    j = 0
    cnt = 0
    while i >= 0 and j < m:
        if matrix[i][j] >= mid:
            cnt += i + 1
            j += 1
        else:
            i -= 1
    return cnt
n, m = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]
k = int(input())
low = a[0][0]
high = a[n - 1][m - 1]
ans = low
while low <= high:
    mid = (low + high) // 2
    if count_ge(a, mid) >= k:
        ans = mid
        low = mid + 1
    else:
        high = mid - 1
print(ans)
```

```python
#W9A3
def smallest_after_one_swap(n):
    s = list(str(n))
    length = len(s)
    last_pos = {}
    for i in range(length):
        last_pos[s[i]] = i
    for i in range(length):
        for d in map(str, range(0, int(s[i]))):
            if d in last_pos and last_pos[d] > i:
                if i == 0 and d == '0':
                    continue
                j = last_pos[d]
                s[i], s[j] = s[j], s[i]
                return int("".join(s))
    return n
n = int(input())
print(smallest_after_one_swap(n))
```

```python
#W9A4
Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    n = len(a)
    m = len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    for i in range(1, n + 1):
        if a[i - 1].islower():
            dp[i][0] = dp[i - 1][0]
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if a[i - 1].islower():
                dp[i][j] = dp[i - 1][j]
                if a[i - 1].upper() == b[j - 1]:
                    dp[i][j] |= dp[i - 1][j - 1]
            else:             
                if a[i - 1] == b[j - 1]:
                    dp[i][j] = dp[i - 1][j - 1]

    print("YES" if dp[n][m] else "NO")
```

```python
#W9A5
def is_subarray(list1, list2):
    n = len(list1)
    m = len(list2)
    if n > m:
        return False
    for i in range(m - n + 1):
        ok = True
        for j in range(n):
            if list2[i + j] != list1[j]:
                ok = False
                break
        if ok:
            return True

    return False
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0 or flowerbed[i - 1] == 0)
            right = (i == n - 1 or flowerbed[i + 1] == 0)
            if left and right:
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True

    return k <= 0
```

```python
#W9A8
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    for m in moves:
        if m == 'R':
            x += 1
        elif m == 'L':
            x -= 1
        elif m == 'U':
            y += 1
        elif m == 'D':
            y -= 1
        if (x, y) in visited:
            return True
        visited.add((x, y))

    return False
```

```python
#W9A9
def Median(nums):
    a, b, c, d, e = nums
    if a < b:
        x1, y1 = a, b
    else:
        x1, y1 = b, a
    if c < d:
        x2, y2 = c, d
    else:
        x2, y2 = d, c
    if x1 < x2:
        low = x1
        mid1 = x2
    else:
        low = x2
        mid1 = x1
    if y1 < y2:
        mid2 = y1
        high = y2
    else:
        mid2 = y2
        high = y1
    if mid1 < mid2:
        if e < mid1:
            return mid1
        elif e < mid2:
            return e
        else:
            return mid2
    else:
        if e < mid2:
            return mid2
        elif e < mid1:
            return e
        else:
            return mid1
```

```python
#W9A10
import bisect
def findRadius(houses, heaters):
    if not houses:
        return 0
    if not heaters:
        return 0  
    houses.sort()
    heaters.sort()
    ans = 0
    for h in houses:
        i = bisect.bisect_left(heaters, h)
        right_dist = float('inf')
        if i < len(heaters):
            right_dist = abs(heaters[i] - h)
        left_dist = float('inf')
        if i > 0:
            left_dist = abs(h - heaters[i - 1])
        closest = min(left_dist, right_dist)

        ans = max(ans, closest)

    return ans
```
