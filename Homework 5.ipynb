```python
#W6A19
def index_of_max(lst):
    max_value = lst[0]
    max_index = 0

    for i in range(1, len(lst)):
        if lst[i] > max_value:
            max_value = lst[i]
            max_index = i

    return max_index
```

```python
#W6A20
def find_first_index(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

```

```python
#W6A23
def remove_less_than_x(lst, x):
    result = []
    for v in lst:
        if v >= x:
            result.append(v)
    return result
```

```python
#W6A24
def move_zero_to_front(lst):
    zeros = []
    others = []

    for x in lst:
        if x == 0:
            zeros.append(x)
        else:
            others.append(x)

    return zeros + others
```

```python
#W6A26
a = list(map(int, input().split()))
n = len(a)
total = 0
for i in range(1, n - 1):
    if a[i] > a[i - 1] and a[i] > a[i + 1]:
        total += a[i]

print(total)
```
