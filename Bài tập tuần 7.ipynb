```python
#W7A1
def binary_search(arr, left, right, target):
    if right >= left:
        mid = left + (right - left) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            return binary_search(arr, left, mid - 1, target)
        else:
            return binary_search(arr, mid + 1, right, target)
    else:
        return -1
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

```python
#W7A4
def most_frequent(arr):
    max_count = 0
    result = None

    for i in range(len(arr)):
        count = 0
        for j in range(len(arr)):
            if arr[j] == arr[i]:
                count += 1

        if count > max_count:
            max_count = count
            result = arr[i]

    return result

```

```python
#W7A5
def count_pairs(arr, X):
    count = 0
    n = len(arr)

    for i in range(n - 1):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == X:
                count += 1

    return count
```

```python
#W7A6
def flatten_list(nested):
    result = []
    for sublist in nested:
        for item in sublist:
            result.append(item)
    return result
```

```python
#W7A7
def length_of_LIS(nums):
    if not nums:
        return 0

    n = len(nums)
    dp = [1] * n

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)
```

```python
#W7A8
import heapq
def min_interval(intervals, queries):
    intervals.sort(key=lambda x: x[0])
    sorted_queries = sorted(enumerate(queries), key=lambda x: x[1])
    res = [-1] * len(queries)
    heap = []  
    i = 0     
    for idx, q in sorted_queries:
        while i < len(intervals) and intervals[i][0] <= q:
            l, r = intervals[i]
            heapq.heappush(heap, (r - l + 1, r))
            i += 1
        while heap and heap[0][1] < q:
            heapq.heappop(heap)
        if heap:
            res[idx] = heap[0][0]
        else:
            res[idx] = -1

    return res
```

```python
#W7A9
def count_smaller(nums):
    result = [0] * len(nums)
    sorted_list = []
    for i in range(len(nums) - 1, -1, -1):
        left, right = 0, len(sorted_list)
        while left < right:
            mid = (left + right) // 2
            if sorted_list[mid] < nums[i]:
                left = mid + 1
            else:
                right = mid

        result[i] = left
        sorted_list.insert(left, nums[i])

    return result
```

```python
#W7A10
def max_subarray_modulo(nums, m):
    prefix = 0
    max_sum = 0
    sorted_prefix = [0]
    for x in nums:
        prefix = (prefix + x) % m
        left, right = 0, len(sorted_prefix)
        while left < right:
            mid = (left + right) // 2
            if sorted_prefix[mid] <= prefix:
                left = mid + 1
            else:
                right = mid

        if left < len(sorted_prefix):
            candidate = (prefix - sorted_prefix[left] + m) % m
            max_sum = max(max_sum, candidate)
        else:
            max_sum = max(max_sum, prefix)
        sorted_prefix.insert(left, prefix)

    return max_sum
```

```python
#W7A11
def count_occurrences(text: str, key: str) -> int:
    count = 0
    i = 0
    n = len(key)

    while i <= len(text) - n:
        if text[i:i+n] == key:
            count += 1
            i += n   
        else:
            i += 1

    return count
```

```python
#W7A13
def longest_harmonious_subsequence(arr):
    freq = {}
    for num in arr:
        freq[num] = freq.get(num, 0) + 1

    max_len = 0
    for num in freq:
        if num + 1 in freq:
            max_len = max(max_len, freq[num] + freq[num + 1])

    return max_len
```
